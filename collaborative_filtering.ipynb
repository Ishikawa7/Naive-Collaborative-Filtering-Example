# Implementation of collaborative filtering 
## user-based recommendation algorithm

```python
import pandas as pd

# Load in the data
df = pd.read_csv('ratings.csv')

# Create a pivot table
df_pivot = df.pivot(index='userId', columns='movieId', values='rating')

# Show the pivot table
df_pivot
```

```python
# calculate correlation matrix between users not including NaN values
corrMatrix = df_pivot.T.corr(method='pearson', min_periods=100, )

# Fill in NaN values in the correlation matrix
corrMatrix = corrMatrix.fillna(0)

# Show the correlation matrix
corrMatrix
```

```python
# Select a user
user = 608

# Set the number of neighbors
k = 10
```

```python
# Get the indices of the nearest neighbors
nearest_neighbors = corrMatrix[user].sort_values(ascending=False).iloc[1:k+1]

# Show the nearest neighbors
nearest_neighbors
```

```python
# Get the ratings of the nearest neighbors
nearest_neighbors_ratings = df_pivot.loc[nearest_neighbors.index]

# Show the ratings of the nearest neighbors
nearest_neighbors_ratings
```

```python
# Get the movies that the user has rated

row = df_pivot.iloc[user]
mask = row.notnull()

rated_movies =  row[mask].index.tolist()
rated_movies
```

```python
len(rated_movies)
```

```python
# filter for movies not rated by the target user
nearest_neighbors_ratings.drop(columns=rated_movies, inplace=True)
nearest_neighbors_ratings
```

```python
# weight rating with correlations
nearest_neighbors_ratings_weighted = nearest_neighbors_ratings.multiply(nearest_neighbors, axis=0)
nearest_neighbors_ratings_weighted
```

```python
# Sum the weighted ratings and normalize
recommendations = nearest_neighbors_ratings_weighted.sum() / nearest_neighbors.sum()
recommendations
```

```python
# Sort the recommendations in descending order
recommendations = recommendations.sort_values(ascending=False)

# Show the top 10 recommendations
recommendations.head(10)
```
